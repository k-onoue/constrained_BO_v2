```python
import os
import re
import pandas as pd


def parse_log_file(filepath):
    import re
    import pandas as pd
    
    # Regex pattern to match lines with mean and std values
    pattern = re.compile(r"UCB value at.*?Mean: ([\d\.-]+), Std: ([\d\.-]+)")
    
    # Initialize an empty list to hold the extracted data
    data = []
    
    # Read the log file line by line
    with open(filepath, 'r') as file:
        for line in file:
            if "sample_independent" in line:
                data.append((None, None))
            else:
                match = pattern.search(line)
                if match:
                    mean = float(match.group(1))
                    std = float(match.group(2))
                    data.append((mean, std))
    
    # Create a DataFrame from the extracted data
    df = pd.DataFrame(data, columns=["Mean", "Std"])
    return df
```

```python
logs_dir = "/Users/keisukeonoue/ws/constrained_BO_v2/results_ackley_7/logs/"

log_files = os.listdir(logs_dir)
log_files
```

```python
file_path = os.path.join(logs_dir, log_files[0]) 
print(file_path)

df = parse_log_file(file_path)
display(df)
```

```python
for log_file in log_files:
    file_path = os.path.join(logs_dir, log_file)
    df = parse_log_file(file_path)
    print(f"File: {log_file}")
    display(df)
    
    # Print only the min and max for Mean and Std columns
    print(df[['Mean', 'Std']].describe().loc[['min', 'max']])
    print("\n\n")
```
